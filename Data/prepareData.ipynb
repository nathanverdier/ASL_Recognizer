```python
df = pd.read_csv('./test/_classes.csv')

# Afficher le nombre de valeurs maximum pour chaque colonne
valMax = 0
valActu = 0
valMin = float('inf')
for col in df.columns[1:]:
    print(f"Colonne {col}:")
    valActu = df[col].value_counts().max()
    print(valActu)
    if valActu > valMax:
        valMax = valActu
    if valActu < valMin:
        valMin = valActu
print("La valeur maximal trouver est :")
print(valMax)
print("La valeur minimal trouver est :")
print(valMin)
```
